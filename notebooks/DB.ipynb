```python
import sqlite3
```

```python
db = sqlite3.connect("../database/BinanceData.db");
```

```python
cursor = db.cursor()
cursor.execute('''SELECT 
    name
FROM 
    sqlite_master 
WHERE 
    type ='table' AND 
    name NOT LIKE 'sqlite_%';''')
```

```python
all_rows = cursor.fetchall()
for row in all_rows:
    print(row)
```
