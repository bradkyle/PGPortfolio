```python
feature_number=3
asset_number=15
window_size=90
```

```python
from oauth2client.client import GoogleCredentials
from googleapiclient import discovery
import random
import numpy
```

```python
PROJECT='axiom-209217'
MODEL='test'
VERSION='test_1'
```

```python
credentials = GoogleCredentials.get_application_default()
api = discovery.build('ml', 'v1', credentials=credentials)
```

```python
prev_w = numpy.random.random((asset_number)).tolist()
inp = numpy.random.random((feature_number,asset_number, window_size))
```

```python
inp.shape[0]
```

```python
# Add your sample instances here.
request_data = {'instances':[
    {
    'input_num':inp.shape[0],
    'previous_w':prev_w,
    'input':inp.tolist()
    }
] }
```

```python
parent = 'projects/%s/models/%s/versions/%s' % (PROJECT, MODEL, VERSION)
response = api.projects().predict(body=request_data, name=parent).execute()
print( "response={0}".format(response))
```
