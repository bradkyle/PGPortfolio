```python
from oauth2client.client import GoogleCredentials
from googleapiclient import discovery
import random
import numpy as np
import requests
```

```python
PROJECT='axiom-209217'
MODEL='test'
VERSION='test_3'
```

```python
credentials = GoogleCredentials.get_application_default()
api = discovery.build('ml', 'v1', credentials=credentials)
```

```python
import json
with open("data_1.json") as f:
    data = json.load(f)
```

```python
print(np.array(data['instances'][0]['input']).shape)
print(np.array(data['instances'][0]['previous_w']).shape)
# print(np.array(np.array(data['instances'][0]['input_num']).shape))
```

```python
parent = 'projects/%s/models/%s/versions/%s' % (PROJECT, MODEL, VERSION)
response = api.projects().predict(body=data, name=parent).execute()
```

```python
output=response['predictions'][0]['output']
```
